# 3. 💻 Code Walkthrough — Line by Line

````python
# ── INTEGERS ──────────────────────────────────────────
age = 25            # Like SQL's INT. Whole numbers only.
year = 2024         # No quotes → Python knows it's a number

# ── FLOATS ────────────────────────────────────────────
salary = 95000.50   # Like SQL's DECIMAL. Has a decimal point.
pi = 3.14159        # R users: same as numeric

# ── STRINGS ───────────────────────────────────────────
name = "Alice"      # Text. Always inside quotes.
city = 'Kolkata'    # Single quotes work too — same thing.

# ── BOOLEANS ──────────────────────────────────────────
is_active = True    # Capital T! Not "true" or "TRUE"
has_account = False # Capital F! R uses TRUE/FALSE; Python uses True/False

# ── THE type() FUNCTION ──────────────────────────────
# This is your x-ray machine — it tells you what's inside
print(type(age))        # → <class 'int'>
print(type(salary))     # → <class 'float'>
print(type(name))       # → <class 'str'>
print(type(is_active))  # → <class 'bool'>
````

````python
# ── KEY PYTHON SUPERPOWER: DYNAMIC TYPING ─────────────
# In SQL you define column types upfront. Python doesn't care.
x = 10          # x is an int right now
x = "hello"     # now x is a string — Python just shrugs
x = 3.14        # now it's a float — this is legal!
# (This freedom can cause bugs — we'll talk about that)

# ── TYPE CONVERSION (like SQL CAST) ───────────────────
number_as_string = "42"        # This looks like a number but it's text
actual_number = int("42")      # Convert: str → int
print(actual_number + 8)       # → 50  ✓

price = 19.99
rounded = int(price)           # Convert float → int (truncates, doesn't round!)
print(rounded)                 # → 19  (not 20!)
````

# 4. 🔨 Mini Project — "Student Profile Card"

````python
# Step 1: Define the student's data using correct types
student_name = "Rahul Sharma"    # str
student_age = 22                 # int
gpa = 8.75                       # float  (out of 10)
is_enrolled = True               # bool
graduation_year = 2026           # int

# Step 2: Python's f-string — your best friend for printing
# Think of it like paste() in R or CONCAT in SQL, but cleaner
print(f"Name: {student_name}")
print(f"Age: {student_age}")
print(f"GPA: {gpa}")
print(f"Enrolled: {is_enrolled}")
print(f"Graduating: {graduation_year}")

# Step 3: Do a calculation inline
years_left = graduation_year - 2024
print(f"Years until graduation: {years_left}")
````

### Common trap — integer division:

> This bites people in binary search problems ALL the time

> mid = (left + right) / 2    # ← WRONG: gives float (e.g. 3.5)

> mid = (left + right) // 2   # ← CORRECT: integer division, gives int (e.g. 3)


## Common trap — None vs 0 vs False:

````python
# These are all "falsy" in Python — they all behave like False in conditions
# But they are NOT the same thing!
x = None   # "nothing here" — like SQL NULL
x = 0      # the number zero
x = False  # boolean false
x = ""     # empty string

# All of these evaluate to False in an if statement
# Interviewers test if you know the difference
````

### Type errors are runtime errors — Python won't warn you at write-time:

````python
age = "25"          # You meant int, wrote str by mistake
result = age + 5    # → TypeError! Crashes at runtime
# In FAANG interviews: runtime crashes = failed test cases
````

# Tool 1 — Multiple assignment (Python superpower):

````python
# Instead of this:
x = 1
y = 2
z = 3

# Python lets you do this:
x, y, z = 1, 2, 3        # same result, one line

# And this (swap without a temp variable — very FAANG):
a, b = 10, 20
a, b = b, a               # now a=20, b=10
print(a, b)               # → 20 10
````

# Tool 2 — Integer division // and modulo %:

````python
17 // 5    # → 3   (how many times does 5 fully fit into 17?)
17 % 5     # → 2   (what's left over?)

# Real example: converting 150 minutes into hours + minutes
total_minutes = 150
hours   = total_minutes // 60    # → 2
minutes = total_minutes % 60     # → 30
# "2 hours and 30 minutes"
````
